```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("./local_data/data_cleaned.csv")
```

```python
data.skew().abs().sort_values(ascending=False).head(50)
```

```python
temp_data = data["number_of_reviews"]
temp_name = "number of reviews"

# Calculate outlier thresholds using IQR method
Q1 = temp_data.quantile(0.25)
Q3 = temp_data.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create figure with appropriate size
fig, ax = plt.subplots(figsize=(10, 8))

# Plot main distribution
sns.histplot(
    data=temp_data[temp_data.between(lower_bound, upper_bound)],
    kde=True,
    stat="density",
    color="#2E86C1",
    alpha=0.6,
    ax=ax,
    label="Normal values",
)

# Highlight outliers in red
sns.histplot(
    data=temp_data[~temp_data.between(lower_bound, upper_bound)],
    kde=False,
    stat="density",
    color="#E74C3C",
    alpha=0.6,
    ax=ax,
    label="Outliers",
)

# Calculate statistical measures
mean_val = temp_data.mean()
median_val = temp_data.median()
skew_val = temp_data.skew()

# Add vertical lines for mean and median
ax.axvline(mean_val, color="#E74C3C", linestyle="--", label=f"Mean: {mean_val:.2f}")
ax.axvline(
    median_val, color="#27AE60", linestyle="--", label=f"Median: {median_val:.2f}"
)

# Customize the plot
ax.set_title(f"Distribution of {temp_name}\nAirbnb Zürich", fontsize=14, pad=20)
ax.set_xlabel(f"{temp_name}", fontsize=12)
ax.set_ylabel("Density", fontsize=12)

# Add statistical annotation
stats_text = f"Skewness: {skew_val:.3f}"


# Customize legend - moved to upper right
ax.legend(
    frameon=True, facecolor="white", edgecolor="none", loc="upper right"
)  # Changed from 'upper left' to 'upper right'


# Add outlier information to stats text
outlier_count = (~temp_data.between(lower_bound, upper_bound)).sum()
outlier_percentage = (outlier_count / len(temp_data)) * 100
stats_text = (
    f"Skewness: {skew_val:.3f}\n"
    f"Outliers: {outlier_count} ({outlier_percentage:.1f}%)\n"
    f"1.5 IQR Range: ${lower_bound:.0f} - ${upper_bound:.0f}"
)
ax.text(
    0.02,
    0.95,
    stats_text,
    transform=ax.transAxes,
    bbox=dict(facecolor="white", alpha=0.8, edgecolor="none"),
    fontsize=10,
    verticalalignment="top",
)


# Customize legend - moved to upper right
ax.legend(
    frameon=True, facecolor="white", edgecolor="none", loc="upper right"
)  # Changed from 'upper left' to 'upper right'

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate correlations with price
correlations = data.corr()["price"].sort_values(ascending=True)
correlations = correlations.drop("price")  # Remove price self-correlation

# Create figure
fig, ax = plt.subplots(figsize=(12, 6))

# Create bar plot
bars = ax.barh(
    range(len(correlations)),
    correlations,
    color=["#E74C3C" if x < 0 else "#2E86C1" for x in correlations],
    alpha=0.6,
)

# Customize the plot
ax.set_title(
    "Correlation of Features with rental Price\nAirbnb Zürich", fontsize=14, pad=20
)
ax.set_xlabel("Correlation Coefficient", fontsize=12)

# Remove yticks and labels
ax.set_yticks([])

# Add gridlines
ax.grid(True, linestyle="--", alpha=0.7)

# Add a vertical line at x=0
ax.axvline(x=0, color="black", linestyle="-", linewidth=0.5)

# Add correlation strength zones
ax.axvline(x=0.5, color="gray", linestyle="--", alpha=0.3)
ax.axvline(x=-0.5, color="gray", linestyle="--", alpha=0.3)
ax.text(0.5, len(correlations), "Strong Positive", ha="left", va="bottom", alpha=0.5)
ax.text(-0.5, len(correlations), "Strong Negative", ha="right", va="bottom", alpha=0.5)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
